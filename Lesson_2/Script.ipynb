```python
# 3. Напишите код на Python, реализующий построение графиков:
#  окружности,
#  эллипса,
#  гиперболы.

def func_circle(x, radius):
    return np.sqrt(radius**2 - x**2)


def func_ellipse(x, a, b):
    return b * np.sqrt(1 - x**2/a**2)


def func_hyperbola(x, a, b):
    return b * np.sqrt(x**2/a**2 - 1)


fig, ax = plt.subplots(nrows=3, ncols=1)
plt.subplots_adjust(hspace=0.40)
radius = 5

x = np.linspace(-5, 5, 1000)
ax[0].set_title('Окружность')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_xlim(-10, 10)
ax[0].plot(x, func_circle(x, radius))
ax[0].plot(x, -func_circle(x, radius))

ax[0].grid(True, which='both')

x = np.linspace(-3, 3, 1000)
ax[1].set_title('Эллипс')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_xlim(-4, 4)
ax[1].plot(x, func_ellipse(x, 3, 2))
ax[1].plot(x, -func_ellipse(x, 3, 2))
ax[1].grid()

x = np.linspace(-5, 5, 1000)
ax[2].set_title('Гипербола')
ax[2].set_xlabel('x')
ax[2].set_ylabel('y')
ax[2].set_xlim(-4, 4)
ax[2].plot(x, func_hyperbola(x, 1, 2))
ax[2].plot(x, -func_hyperbola(x, 1, 2))
ax[2].grid()

fig.set_figheight(14)
plt.show()
```

```python
# 5. Нарисуйте трехмерный график двух параллельных плоскостей.

fig = figure()
ax = Axes3D(fig)

X = np.arange(-10, 10, 10)
Y = np.arange(-50, 50, 20)
X, Y = np.meshgrid(X, Y)

Z1 = 2*X + 3*Y + 20
Z2 = 2*X + 3*Y

ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
# Нарисуйте трехмерный график двух любых поверхностей второго порядка.

fig = figure()
ax = Axes3D(fig)
X = np.arange(-50, 50, 1)
Y = np.arange(-50, 50, 1)
X, Y = np.meshgrid(X, Y)

Z1 = 4 - X*X - Y*Y
Z2 = 2*X*X - 6*Y*Y
ax.plot_surface(X, Y, Z1)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
# 6.Напишите код, который будет переводить полярные координаты в декартовы.

def convert_to_cartesian_coordinates(alfa, radius):
    x = radius * np.cos(alfa)
    y = radius * np.sin(alfa)
    return x, y


x1, y1 = convert_to_cartesian_coordinates(np.pi, 3)
print(x1, " ", y1)
```

```python
# Напишите код, который будет рисовать график окружности в полярных координатах.

rads = np.arange(0, (2*np.pi), 0.01)
plt.axes(projection='polar')

for radian in rads:
    plt.polar(radian, 3, 'o')
plt.show()
```

```python
# Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.
radius = np.arange(0, 10, 0.1)
plt.axes(projection='polar')
for r in radius:
    plt.polar(np.pi/4, r, 'o')
plt.show()
```
